Hãy thực hiện các yêu cầu sau:
<ol>
    <li>In list thông tin ra màn hình dưới dạng bảng.</li>
    <li>Tìm và in thông tin vận động viên giàu thành tích nhất. Biết rằng thứ hạng thành tích của vận động viên là số huy chương giành được theo thứ tự ưu tiên vàng, bạc, đồng.</li>

    <li>Thống kê số lượng huy chương từng loại (vàng, bạc, đồng) của từng quốc gia có vận động viên tham gia. Kết quả được lưu trong một từ điển có cấu trúc tương tự như yêu cầu trước.</li>

````python
infos = [
    {'name': 'L. Marchand', 'nation': 'France', 'programme': 'Swimming', 'gold': 4, 'silver': 0, 'bronze': 1},
    {'name': 'T. Huske', 'nation': 'US', 'programme': 'Swimming', 'gold': 3, 'silver': 2, 'bronze': 0},
    {'name': 'M. OCallaghan', 'nation': 'Australia', 'programme': 'Swimming', 'gold': 3, 'silver': 1, 'bronze': 1},
    {'name': 'S. Biles', 'nation': 'Canada', 'programme': 'Artistic gymnastics', 'gold': 3, 'silver': 1, 'bronze': 0},
    {'name': 'M. McIntosh', 'nation': 'France', 'programme': 'Swimming', 'gold': 3, 'silver': 1, 'bronze': 0},
    {'name': 'S. Oka', 'nation': 'Japan', 'programme': 'Artistic gymnastics', 'gold': 3, 'silver': 0, 'bronze': 1},
    {'name': 'G. Thomas', 'nation': 'US', 'programme': 'Athletics', 'gold': 3, 'silver': 0, 'bronze': 0},
    {'name': 'H. Lavreysen', 'nation': 'Netherlands', 'programme': 'Track cycling', 'gold': 3, 'silver': 0, 'bronze': 0},
    {'name': 'W. J. Kim', 'nation': 'South Korea', 'programme': 'Archery', 'gold': 3, 'silver': 0, 'bronze': 0},
    {'name': 'S. H. Lim', 'nation': 'South Korea', 'programme': 'Archery', 'gold': 3, 'silver': 0, 'bronze': 0}
]

def print_all():
    print("{:<15} {:<13} {:<21} {:<7} {:<9} {:<7}".format("NAME:", "NATION:", "PROGRAMME:", "GOLD:", "SILVER:", "BRONZE:"))
    for info in infos:
        print("{:<15} {:<13} {:21} {:<7} {:<9} {:<7}".format(info["name"], info["nation"], info["programme"], info["gold"], info["silver"], info["bronze"]))
    print("\n")

def find_best_athlete():
    most_gold = []
    gold_counter = -1
    
    for info in infos:
        if info["gold"] > gold_counter:
            gold_counter = info["gold"]
            most_gold.clear()
            most_gold.append(info["name"])
        elif info["gold"] == gold_counter:
            most_gold.append(info["name"])    
        
    print("BEST ATHLETE:")
    for info in infos:
        if info["name"] in most_gold:
            print(f"NAME: {info['name']}, NATION: {info['nation']}, PROGRAMME: {info['programme']}, GOLD: {info['gold']}, SILVER: {info['silver']}, BRONZE: {info['bronze']}")
    print("\n")

def find_best_nation():
    nation_medals = {}
    
    for info in infos:
        nation = info['nation']
        if nation not in nation_medals:
            nation_medals[nation] = {'gold': 0, 'silver': 0, 'bronze': 0}
        
        nation_medals[nation]['gold'] += info['gold']
        nation_medals[nation]['silver'] += info['silver']
        nation_medals[nation]['bronze'] += info['bronze']
    
    best_nation = max(nation_medals.items(), key=lambda x: (x[1]['gold'], x[1]['silver'], x[1]['bronze']))
    print("BEST NATION:")
    print(f"Nation: {best_nation[0]}, Gold: {best_nation[1]['gold']}, Silver: {best_nation[1]['silver']}, Bronze: {best_nation[1]['bronze']}\n")

def sort_by_total_medals():
    sorted_infos = sorted(infos, key=lambda x: (x['gold'] + x['silver'] + x['bronze']), reverse=True)
    print("Sorted Athletes by Total Medals:")
    for info in sorted_infos:
        print(f"{info['name']} - Total Medals: {info['gold'] + info['silver'] + info['bronze']}")
    print("\n")

def medal_statistics_by_programme():
    programme_stats = {}
    
    for info in infos:
        programme = info['programme']
        if programme not in programme_stats:
            programme_stats[programme] = {'gold': 0, 'silver': 0, 'bronze': 0}
        
        programme_stats[programme]['gold'] += info['gold']
        programme_stats[programme]['silver'] += info['silver']
        programme_stats[programme]['bronze'] += info['bronze']
    
    print("Medal Statistics by Programme:")
    for programme, stats in programme_stats.items():
        print(f"{programme}: Gold: {stats['gold']}, Silver: {stats['silver']}, Bronze: {stats['bronze']}")
    print("\n")

def medal_statistics_by_nation():
    nation_stats = {}
    
    for info in infos:
        nation = info['nation']
        if nation not in nation_stats:
            nation_stats[nation] = {'gold': 0, 'silver': 0, 'bronze': 0}
        
        nation_stats[nation]['gold'] += info['gold']
        nation_stats[nation]['silver'] += info['silver']
        nation_stats[nation]['bronze'] += info['bronze']
    
    print("Medal Statistics by Nation:")
    for nation, stats in nation_stats.items():
        print(f"{nation}: Gold: {stats['gold']}, Silver: {stats['silver']}, Bronze: {stats['bronze']}")
    print("\n")

print_all()
find_best_athlete()
find_best_nation()
sort_by_total_medals()
medal_statistics_by_programme()
medal_statistics_by_nation()
````
